```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV, learning_curve
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Cargar el conjunto de datos
df = pd.read_csv('/content/energy_dataset_.csv', sep=',')

# Verificar y limpiar nombres de columnas
df.columns = df.columns.str.strip()

# Realizar anlisis exploratorio de datos
print(df.head())
print(df.describe())

# Visualizar la distribucin de las clases
sns.countplot(x='Type_of_Renewable_Energy', data=df)
plt.title('Distribucin de Clases')
plt.show()

# Visualizar la distribucin de las variables numricas
df.hist(bins=50, figsize=(20, 15))
plt.show()

# Visualizar la correlacin entre las variables
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```
